## Removing the nan values 

```python
df.dropna(axis=0, inplace=True)
```

## mapping the values of validation status column

```python
df['Validation Status']=df['Validation Status'].map({'Accepted':1,'Rejected':0,'Information Required':2,'Escalate':3,'Pending':4})
```

```python
df.head(2)
```

```python
df1=df.copy() # Copying dataframe from df to df1 
```

## wards and validation status - approval

```python
val=df1.groupby(['Validation Status'])
```

```python
va_approval=val.get_group(1)
```

```python
va_approval.head(2)
```

```python
grouped = va_approval.groupby(['Ward'])['Validation Status'].sum()
```

```python
val1=grouped.to_frame()
```

## Arrange the wards in a descending order based upon the number of approvals obtained in each ward

```python
## solution
val1.sort_values(by=['Validation Status'],ascending=False)
```

## localities  vs validation status - approval

```python
grouped1=va_approval.groupby(['Ward','Localities'])['Validation Status'].sum()
```

```python
val2=grouped1.to_frame()
val2
```

## Localities with approval status

```python
val2.sort_values(by=['Validation Status'],ascending=False)
```

## Create a visualisation which shows the breakdown of approvals and rejections in the top 10 localities and map them to their respective wards in the same visualisation.
